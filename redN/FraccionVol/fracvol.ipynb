# Leyendolos en y juntandolos en una misma base

```python
files = [file for file in os.listdir('./vol_export')]

vol = pd.DataFrame()
for file in files:
  df = pd.read_csv('./vol_export/'+file)
  vol = pd.concat([vol, df], axis= 0)
```

```python
files = [file for file in os.listdir('./nuevo_vol')]

vol_new = pd.DataFrame()
for file in files:
  df = pd.read_csv('./nuevo_vol/'+file)
  vol_new = pd.concat([vol, df], axis= 0)
```
