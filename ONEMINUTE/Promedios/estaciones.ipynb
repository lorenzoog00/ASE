```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Estaciones por separado

```python
files = [file for file in os.listdir('./invierno')]
invierno = {}
for file in files:
    invierno[file] =  pd.read_csv('./invierno/'+ file)
```

```python
files = [file for file in os.listdir('./verano')]
verano = {}
for file in files:
    verano[file] =  pd.read_csv('./verano/'+ file)
```

```python
files = [file for file in os.listdir('./primavera')]
primavera = {}
for file in files:
    primavera[file] =  pd.read_csv('./primavera/'+ file)
```

```python
files = [file for file in os.listdir('./otono')]
otono = {}
for file in files:
    otono[file] =  pd.read_csv('./otono/'+ file)
```

# Invierno

```python
plt.style.use('ggplot')
plt.figure(figsize= (20,10))
plt.title('Radiación por hora en invierno')
plt.xlabel('Minuto del día')
plt.ylabel('Radiación promedio de todos los días del mes')

for bowser in invierno:
    plt.plot(invierno[bowser].TOA5, invierno[bowser]['Radiación Promedio'], label = bowser )

plt.xticks(invierno['final diciembre22.csv_output.csv']['Unnamed: 0'][::120])
plt.legend()
```
