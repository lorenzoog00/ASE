```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
files = [file for file in os.listdir('./CSVmeses')]
luigi = {}
for file in files:
    luigi[file] =  pd.read_csv('./CSVmeses/'+ file)
```

# Extrayendo promedios de temperaturas, presiones y rediacion de cada a√±o

```python
promedios_temp = []
promedios_rad = []
promedios_pres = []
```

```python

for daisy in luigi:
    promedios_pres.append(luigi[daisy]['CPU:Est_meteo1.CR1X'].mean())
    promedios_temp.append(luigi[daisy]['8543.1'].mean())
    promedios_rad.append(luigi[daisy]['CR1000X'].mean())
```

```python
x = files
y = promedios_temp
plt.figure(figsize= (20,10))
plt.title('Temperatura Junio 2021-2022 Mayo')
plt.bar(x,y)
plt.show()
```

```python
x = files
y = promedios_pres
plt.figure(figsize= (20,10))
plt.title('Presi√≥n Junio 2021-2022 Mayo')
plt.bar(x,y)
plt.show()
```

```python
x = files
y = promedios_rad
plt.figure(figsize= (20,10))
plt.title('Radiaci√≥n Junio 2021-2022 Mayo')
plt.bar(x,y)
plt.show()
```
