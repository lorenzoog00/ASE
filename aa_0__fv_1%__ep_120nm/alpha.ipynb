```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = [file for file in os.listdir('./alphas')]
mario = {}
for file in files:
    mario[file] =  pd.read_csv('./alphas/'+ file)
```

# Renombrando los nanometros y generando graficas para cada uno

```python
for df in mario:
    mario[df].columns =['nm', 'refraccin']

    plt.plot(mario[df]['nm'], mario[df]['refraccin'])
    plt.xlabel('nm')
    plt.ylabel('refraccin')
    plt.title(str(df))
    max_value = mario[df]['refraccin'].max()
    max_index = mario[df]['nm'].idxmax()
    plt.scatter(x=max_index, y=max_value, color='r', label='Max Value')
    plt.annotate(str(max_value), xy=(max_index, max_value))
    plt.show()
```
