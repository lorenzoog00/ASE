```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
files = [file for file in os.listdir('./alphas')]
mario = {}
for file in files:
    mario[file] =  pd.read_csv('./alphas/'+ file)
```

# Renombrando los nanometros y generando graficas para cada uno

```python
for df in mario:
    mario[df].columns =['longitud de onda', 'coeficiente de extinción']

    plt.plot(mario[df]['longitud de onda'], mario[df]['coeficiente de extinción'])
    plt.xlabel('longitud de onda')
    plt.ylabel('coeficiente de extinción')
    plt.title(str(df))
    max_value = mario[df]['coeficiente de extinción'].max()
    max_index = mario[df]['longitud de onda'].idxmax()
    plt.scatter(x=max_index, y=max_value, color='r', label='Max Value')
    plt.annotate(str(max_value), xy=(max_index, max_value))
    plt.show()
```
